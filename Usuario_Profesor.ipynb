```python
class SistemaGestionAcademica:
    def __init__(self):
        self.profesores = {}
        self.estudiantes = {}

    def iniciar_sesion_profesor(self, nombre_usuario, contrasena):
        # Verificar las credenciales del profesor y permitir el inicio de sesión
        # Si las credenciales son válidas, establecer al profesor como usuario activo en el sistema
        pass

    def ingresar_notas(self, profesor_id, asignatura, notas):
        # Verificar si el profesor está autorizado y activo en el sistema
        if profesor_id in self.profesores:
            # Verificar si la asignatura existe para el profesor
            if asignatura in self.profesores[profesor_id]["asignaturas"]:
                # Asociar las notas ingresadas a la asignatura correspondiente
                self.profesores[profesor_id]["asignaturas"][asignatura]["notas"] = notas
                # Calcular el promedio de las notas ingresadas
                promedio = sum(notas) / len(notas)
                # Actualizar el promedio de la asignatura
                self.profesores[profesor_id]["asignaturas"][asignatura]["promedio"] = promedio
                # Actualizar las notas en el perfil de cada estudiante correspondiente
                for estudiante_id in self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"]:
                    self.estudiantes[estudiante_id]["asignaturas"][asignatura]["notas"] = notas
                    self.estudiantes[estudiante_id]["asignaturas"][asignatura]["promedio"] = promedio
                # Mostrar un mensaje de éxito
                print("Notas ingresadas correctamente.")
            else:
                print("La asignatura no existe para el profesor.")
        else:
            print("El profesor no está autorizado o no ha iniciado sesión.")

    def modificar_notas(self, profesor_id, asignatura, estudiante_id, notas):
        # Verificar si el profesor está autorizado y activo en el sistema
        if profesor_id in self.profesores:
            # Verificar si la asignatura existe para el profesor
            if asignatura in self.profesores[profesor_id]["asignaturas"]:
                # Verificar si el estudiante está inscrito en la asignatura
                if estudiante_id in self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"]:
                    # Actualizar las notas del estudiante
                    self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"][estudiante_id] = notas
                    # Recalcular el promedio de la asignatura
                    notas_estudiantes = list(self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"].values())
                    promedio = sum(notas_estudiantes) / len(notas_estudiantes)
                    # Actualizar el promedio de la asignatura
                    self.profesores[profesor_id]["asignaturas"][asignatura]["promedio"] = promedio
                    # Actualizar las notas en el perfil del estudiante correspondiente
                    self.estudiantes[estudiante_id]["asignaturas"][asignatura]["notas"] = notas
                    self.estudiantes[estudiante_id]["asignaturas"][asignatura]["promedio"] = promedio
                    # Mostrar un mensaje de éxito
                    print("Notas modificadas correctamente.")
                else:
                    print("El estudiante no está inscrito en la asignatura.")
            else:
                print("La asignatura no existe para el profesor.")
        else:
            print("El profesor no está autorizado o no ha iniciado sesión.")

    def agregar_profesor(self, profesor_id, nombre_usuario, contrasena):
        # Agregar un nuevo profesor al sistema con sus credenciales
        self.profesores[profesor_id] = {
            "nombre_usuario": nombre_usuario,
            "contrasena": contrasena,
            "asignaturas": {}
        }

    def agregar_estudiante(self, estudiante_id, nombre):
        # Agregar un nuevo estudiante al sistema
        self.estudiantes[estudiante_id] = {
            "nombre": nombre,
            "asignaturas": {}
        }

    def inscribir_estudiante_asignatura(self, profesor_id, asignatura, estudiante_id):
        # Verificar si el profesor está autorizado y activo en el sistema
        if profesor_id in self.profesores:
            # Verificar si la asignatura existe para el profesor
            if asignatura in self.profesores[profesor_id]["asignaturas"]:
                # Verificar si el estudiante ya está inscrito en la asignatura
                if estudiante_id not in self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"]:
                    # Agregar al estudiante a la lista de estudiantes de la asignatura
                    self.profesores[profesor_id]["asignaturas"][asignatura]["estudiantes"][estudiante_id] = []
                    # Agregar la asignatura al perfil del estudiante
                    self.estudiantes[estudiante_id]["asignaturas"][asignatura] = {
                        "notas": [],
                        "promedio": None
                    }
                    # Mostrar un mensaje de éxito
                    print("Estudiante inscrito correctamente en la asignatura.")
                else:
                    print("El estudiante ya está inscrito en la asignatura.")
            else:
                print("La asignatura no existe para el profesor.")
        else:
            print("El profesor no está autorizado o no ha iniciado sesión.")
```
